```python
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt

from src.data.unzipping_files import unzip_files
```

###  Unzip files

```python
DATASET = 'CT-ORG'
SOURCE_FOLDER = f'../data/raw/{DATASET}/'
DEST_FOLDER = f'../data/processed/{DATASET}/'

unzip_files(SOURCE_FOLDER, DEST_FOLDER)
```

### Read file

```python
raw_test_data = nib.load('../data/processed/CT-ORG/volume-0.nii')
test_data = raw_test_data.get_fdata()
print(f'Shape {test_data.shape} \nValues: {np.unique(test_data, return_counts=True)}')
```

### Visualize

```python
NUM_IMAGES = 10

plt.figure(figsize=(20, 30))
for i in range(1, NUM_IMAGES):
    plt.subplot(1, NUM_IMAGES, i)
    plt.imshow(test_data[:, :, i])
```
